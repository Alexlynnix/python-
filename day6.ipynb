```python
counts = []
for i, j in dict_arr.items():
    counts.append(j)
print(counts)
```

```python
d_max = counts[0]
for i in counts:
    if i > d_max:
        d_max =i
print(d_max)
```

### using a while loop to build a guess game


```python
guess_here = 10
guess_count = 0
number_guess = 3

while guess_count < number_guess:
    guess = int(input("enter guess: "))
    guess_count = guess_count + 1
    if guess_here == guess:
        print("your guess was right")
    else:
        print("guess again!")
else:
        print("game over!!!")
```

```python
dict_data = [{"name":"tobi", "scores":{"eng":55,
                                      "math":75,
                                      "econs":95},
             "class":"ss1",
             "family":"both"},
            {"name":"manuel", "scores":{"eng":44,
                                      "math":65,
                                      "econs":79},
             "class":"ss2",
             "family":"single"}]
print(dict_data)


```

```python
1. find the average scores for each of the student
2. add that average score inside the dictionary i.e: "average" : 50
3. write a query that will select all students in a. ss2
b. both
4. what is the modal score for tobi
```

```python
new_scores = []
for i in dict_data:
    arr = []
    score_vals = i["scores"]
    for n, j in score_vals.items():
        arr.append(j)
    new_scores.append(arr)
print(new_scores)
    
```

```python
from statistics import mean

for i in range(len(new_scores)):
    avg = mean(new_scores[i])
    print(round(avg),2)
    dict_data[i]["avg_score"] = round(avg, 2)
print(dict_data)

    
```

```python
for i in range (1, len(new_scores)):
    avg_score = round(sum(new_scores[i])/len(new_scores[i]), 2)
    dict_data[i]["avg_score"] = avg_score
print(dict_data)
```
