## lambda functions

```python
def add_val(x):
    return(x+1)
add_val(2)
```

```python
add_val = lambda x:x+1
add_val(2)
```

```python
vamp_list = [13, 90, 17, 59, 21, 60, 5]
a = lambda x : (x > 18)
new_list = list(filter(a,vamp_list))
print(new_list)
```

```python
names = ["john", "rihad", "perez", "keita"]
a = lambda x : (x[0]).upper() + x [1:]
new_names = list(map(a, names))
print(new_names)
```

```python
# importing the file "online Retail"
import os
import pandas as pd
file = "C:/Users/alexl/Documents/Python Scripts/Online Retail.xlsx"
df = pd.read_excel(file)
df.head()
```

```python
missing = df.isna().sum().reset_index()
missing.columns = ["columns", "counts"]
missing
```

```python
import numpy as np

nan = missing[missing["counts"] > 0]
msn = np.array(nan["columns"])
msn
```

```python
for i in msn:
    new_df = df.fillna("missing")
```

```python
new_df.isna().sum()
```

```python
new_df.head()
```

```python
new_df["InvoiceDate"] = new_df["InvoiceDate"].astype(str)
```

```python
new_df.info()
```

```python
new_df[["invoice_date", "invoice_time"]] = new_df["InvoiceDate"].str.split(" ", expand= True)
```

```python
new_df.head()
```

```python
# the product that sold the highest in quantity
# highest_sold = new_df["Quantity"].value_counts().reset_index()
# highest_sold.columns = ["Quantity", "counts"]
# highest_sold
```

```python
qts = []
qty = new_df["Quantity"]
for i in qty:
    v = abs(i)
    qts.append(v)
```

```python
qts
```

```python
new_df = pd.concat([new_df, pd.DataFrame(qts)], axis =1)
```

```python
new_df.head()
```

```python
new_df = new_df.drop(columns = "Quantity", axis =1)
```

```python
name = {0:"Quantity"}
new_df.rename(columns = name, inplace=True)
new_df.columns
```

```python
new_df["Quantity"].value_counts
```

```python
most_sold = new_df[["Description", "Quantity", "Country"]]
mst = most_sold.groupby(["Description","Country"])["Quantity"].sum().reset_index()
mst.columns = ["Products", "Country", "Total_Qty"]

mst.loc[mst["Total_Qty"] == max(mst["Total_Qty"])]
# avg_tr(kind = "bar", color = "purple")
```

```python
new_df["InvoiceDate"] = pd.to_datetime(new_df["InvoiceDate"])
new_df["Date"] = pd.to_datetime(new_df["invoice_date"])
new_df["Time"] = pd.to_datetime(new_df["invoice_time"])
```

```python
new_df.info()
```

```python
new_df["Hour"] = new_df["Time"].dt.hour
new_df["Minutes"] = new_df["Time"].dt.minute
```

```python
goods = new_df.groupby("Hour")["Quantity"].sum()
goods.sort_values(ascending = False)
```

```python
night = new_df[(new_df.Hour >= 19)]
n = night.groupby("Country")["Quantity"].sum()
n.sort_values(ascending = False)
```
