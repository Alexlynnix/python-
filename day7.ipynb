```python
dict_data = [{"name":"tobi", "scores":{"eng":55,
                                      "math":75,
                                      "econs":95},
             "class":"ss1",
             "family":"both"},
            {"name":"manuel", "scores":{"eng":44,
                                      "math":65,
                                      "econs":79},
             "class":"ss2",
             "family":"single"}]
print(dict_data)
```

```python
import statistics as stat

new_scores = []
for i in dict_data:
    arr = []
    score_vals = i["scores"]
    for n, j in score_vals.items():
        arr.append(j)
    new_scores.append(arr)
print(new_scores)
```

```python
def scores(dict_data):
    new_scores = []
    for i in dict_data:
        arr = []
        score_vals = i["scores"]
        for n, j in score_vals.items():
            arr.append(j)
        new_scores.append(arr)
    return new_scores


def avg_score(new_scores):
    mean = []
    for i in new_scores:
        a = round(stat.mean(i), 2)
        mean.append(a)
    return mean

scores = scores(dict_data)
avg = avg_score(scores)
print(avg)
```
